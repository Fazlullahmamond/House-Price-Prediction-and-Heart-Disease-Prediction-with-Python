# importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

# Loading Dataset

```python
data = pd.read_csv('USA_Housing.csv')
data.head()
```

# Drop Address Column

```python
data = data.drop(['Address'], axis=1)
data.head()
```

# checking for missing data

sns.heatmap(data.isnull())

```python
plt.scatter(data['Avg. Area House Age'], data['Price'])
plt.xlabel('Avg. Area House Age')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['Avg. Area Number of Rooms'], data['Price'])
plt.xlabel('Avg. Area Number of Rooms')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['Area Population'], data['Price'])
plt.xlabel('Area Population')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['Avg. Area Income'], data['Price'])
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['Price'], data['Avg. Area Number of Bedrooms'])
plt.xlabel('Price')
plt.ylabel('Avg. Area Number of Bedrooms')
plt.show()
```

```python
# Train test split
```

```python
X = data.drop(['Price'], axis=1)
Y = data['Price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.30)
```

# Training and Predicting

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
a = model.predict(np.array([79545.4, 5.682861, 7, 4, 23086]).reshape(1, -1))
print("prdicted price is $"+str(round(a[0])))
```

```python
predictions = model.predict(X_test)
predictions
```
